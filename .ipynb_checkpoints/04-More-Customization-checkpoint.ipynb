Subplots

```python
plt.figure(figsize=(8,6))
y=np.random.randn(90)
axes1=plt.subplot2grid((3,3),(0,0),colspan=3)
axes1.plot(y)
axes2=plt.subplot2grid((3,3),(1,0),colspan=2)
axes2.hist(y,bins=20,color='r')
axes3=plt.subplot2grid((3,3),(1,2))
axes4=plt.subplot2grid((3,3),(2,0))
axes5=plt.subplot2grid((3,3),(2,1),colspan=2)
all_axes=plt.gcf().axes
for ax in all_axes:
    for ticklabel in ax.get_xticklabels()+ax.get_yticklabels():
        ticklabel.set_fontsize(5)
plt.suptitle("Demo of subplot2grid")
plt.show()
```

contour line

```python
def process_signals(x,y):
    return(1-(x**2+y**2)*np.exp(-y**3/3))
x=np.arange(-1.5,1.5,0.05)
y=np.arange(-1.5,1.5,0.05)
X,Y=np.meshgrid(x,y)
Z=process_signals(X,Y)
N=np.arange(-1,1.5,0.3)
CS=plt.contour(Z,N,linewidths=2,cmap=mpl.cm.jet)
plt.clabel(CS,inline=True,fmt='%1.1f',fontsize=10)
plt.colorbar(CS)
plt.title('My signal Function')
```

add colors under ploted line

```python
t=np.arange(-1,1,0.01)
y=[np.exp(i) for i in t]
plt.plot(t,y,color='blue',lw=2)
plt.fill_between(t,y,color='silver')
plt.xlim([-1,1])
plt.show()
```

polar plot

```python
figsize=8
colormap=lambda r: plt.cm.Set2(2/20.)
N=18
fig=plt.figure(figsize=(figsize,figsize))
ax=fig.add_axes([0.2,0.2,0.7,0.7],polar=True) # init a polar plot
theta=np.arange(0,2*np.pi,2*np.pi/N)
radii=20*np.random.rand(N)
width=np.pi/3*np.random.rand(N)
bars=ax.bar(theta,radii,width=width,bottom=1.0)
for r, bar in zip(radii,bars):
    bar.set_facecolor(colormap(r))
    bar.set_alpha(0.6)
    
```
