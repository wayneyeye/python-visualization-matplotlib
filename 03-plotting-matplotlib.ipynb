```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

```python
x=[1,2,3,4]
y=[5,4,3,2]
plt.figure()
plt.subplot(231)
plt.plot(x,y)
plt.subplot(232)
plt.bar(x,y)
plt.subplot(233)
plt.barh(x,y)

plt.subplot(234)
plt.bar(x,y)
y1=[7,8,5,3]
plt.bar(x,y1,bottom=y,color='r')
plt.subplot(235)
plt.boxplot(x)
plt.subplot(236)
plt.scatter(x,y)
```

Sine and Cosine Curve

```python
x=np.linspace(-np.pi,np.pi,256,endpoint=True)
y=np.cos(x)
y1=np.sin(x)
plt.plot(x,y)
plt.plot(x,y1)
plt.title("Functions $\sin$ and $\cos$") # set title
plt.xlim(-3,3)#set x limit
#format x ticks
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-0.5\pi$',r'$0$',r'$+\pi/2$',r'$+\pi$'])
```

legends and label

```python
x1=np.random.normal(30,3,100)
x2=np.random.normal(20,2,100)
x3=np.random.normal(10,3,100)
plt.plot(x1,label='plot')
plt.plot(x2,label='second plot')
plt.plot(x3,label='3rd plot')
plt.ylim([0,50])
plt.legend(bbox_to_anchor=(0,1.02,1,0.102),loc=3,ncol=3,mode="expand",borderaxespad=0)
plt.annotate("important value",(55,20),xycoords='data',xytext=(5,42),arrowprops=dict(arrowstyle='->'))
```

histogram

```python
mu=100
sigma=15
x=np.random.normal(mu,sigma,100)
```

```python
ax=plt.gca()
ax.hist(x,bins=35,color='r')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
```

piechart

```python
plt.figure(1,figsize=(6,6))
ax=plt.axes([0.1,0.1,0.8,0.8])
x=[15,30,35,10]
explode=(0.1,0.1,0.1,0.1)
plt.pie(x,labels=('Spring','Summer','Fall','Winter'),explode=explode,autopct='%1.1f%%',startangle=67)
```

filling colors in plot

```python
x=np.arange(0,2,0.01)
y1=np.sin(2*np.pi*x)
y2=1.2*np.sin(4*np.pi*x)
fig=plt.figure()
ax=plt.gca()
# ax.plot(x,y1,x,y2,color='k')
#filling colors
ax.fill_between(x,y1,y2,where=y2>=y1,facecolor='darkblue',interpolate=True) # where accepts a boolean
ax.fill_between(x,y1,y2,where=y2<=y1,facecolor='deeppink',interpolate=True)
```

colored scatter plot

```python
x=np.random.randn(1000)
y1=np.random.randn(len(x))
y2=1.2+np.exp(x)
```

```python
ax1=plt.subplot(121)
plt.scatter(x,y1,color='indigo',alpha=0.3,edgecolors='white',label='no correlation')
ax1=plt.subplot(122)
plt.scatter(x,y2,color='green',alpha=0.3,edgecolors='white',label='strong correlation')
```
